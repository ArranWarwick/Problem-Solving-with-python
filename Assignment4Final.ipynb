```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

A=np.array([1,1/2,3/4])
for i in np.arange(3,25):
    a=1-(1/2)*A[i-1]
    A=np.append(A,a)
plt.plot(A)
plt.grid('on')
plt.show()
```

```python
import numpy as np

def solve_recurrence_relation(n):
    p = np.zeros(n+1)  # Create an array to store the values of p_n
    p[0] = 1  # Initial value for p_0
    
    for i in range(1, n+1):
        p[i] = 1 - (1/2)*p[i-1]  # Calculate the value of p_n using the recurrence relation
    
    return p

n = 1000# Define the value of n
result = solve_recurrence_relation(n)
print(result)
```
