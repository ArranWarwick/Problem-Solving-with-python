```python
import random
import matplotlib.pyplot as plt

def lastnumber(n):
    hat = list(range(1, n + 1))
    while len(hat) > 1:
        x, y = random.sample(hat, 2)
        difference = abs(x - y)
        hat.remove(x)
        hat.remove(y)
        hat.append(difference)
    return hat[0]

def runs(n, k):
    results = []
    for _ in range(k):
        result = lastnumber(n)
        results.append(result)
    
    plt.figure(figsize=(10, 6))
    bins = range(0, n + 2)  # Including 0 in the bins
    freq, bins, patches = plt.hist(results, bins=bins, edgecolor='black', align='left')

    for freq, patch in zip(freq, patches):
        plt.text(patch.get_x() + patch.get_width() / 2, patch.get_height(), 
                 f'{int(freq)}', ha='center', va='bottom')

    plt.title('Frequency of Results')
    plt.xlabel('Result')
    plt.ylabel('Frequency')
    plt.xticks(range(0, n + 1))
    plt.show()

    return results

n = int(input("Enter a positive integer for 'n': "))
k = int(input("Enter how many times to run 'lastnumber' for 'k': "))
results = runs(n, k)
print(f"The results of {k} runs are: {results}")
```
